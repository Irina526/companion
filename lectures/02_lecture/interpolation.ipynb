```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": ["Helvetica Neue"],
    "font.size": 28,
    })
np.random.seed(6)  # Make problem repetible
```

```python
# Define vector of points
x_vals = np.array([-1.1, -0.4, 0.1, 0.8])
b_vals = np.random.rand(4)  # Random values
```

```python
# Construct linear system
A = np.array([[1, s, s**2, s**3] for s in x_vals])
b = b_vals
```

```python
# Compute coefficients
c = np.linalg.solve(A, b)
c
```

```python
# Evaluate polynomial
def p(x, c):
    return np.sum([c[i] * (x**(i)) for i in range(4)])
x_plot = np.linspace(-1.1, 0.8)
p_plot = np.array([p(x, c) for x in x_plot])
```

```python
# Plot polynomial
fig = plt.figure(figsize=(16, 9))
plt.plot(x_plot, p_plot)
plt.scatter(x_vals, b_vals)
plt.xlabel(r"$x$")
plt.ylabel(r"$p(x)$")
plt.savefig("interp.pdf")
```
